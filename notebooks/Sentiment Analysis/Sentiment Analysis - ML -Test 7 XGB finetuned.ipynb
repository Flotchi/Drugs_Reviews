```python
from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline,Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import recall_score
```

```python
import pandas as pd
import numpy as np
import logic.processing as lp
from sklearn.model_selection import GridSearchCV
```

```python
df = lp.load_data('drugsComTrain_raw.csv')
```

```python
df1 = lp.data_filter(df)
```

```python
df2 = lp.preproc(df1)
```

```python
X = df2['clean']
y = df2['sentiment']
```

```python
Xb, yb = lp.balance_dataset(X,y)
```

# XGB

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor, XGBClassifier
```

```python
pipe = make_pipeline(TfidfVectorizer(),
                     XGBClassifier(objective='binary:logistic'))
```

```python
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('XGB', XGBClassifier(objective='binary:logistic'))
])
```

# XGB Fine Tuning


```python
param_grid = {
    #'tfidf__max_df': [0.75, 1.0],  
    #'tfidf__min_df': [1, 2, 5],
    'tfidf__ngram_range': [(1,1),(1,2)],
    #'xgbclassifier__n_estimators': [50, 100, 200],
    #'xgbclassifier__max_depth': [3, 5, 7],
    'XGB__learning_rate': [0.01, 0.1],
    #xgbclassifier__subsample': [0.6, 0.8, 1.0],
    #'xgbclassifier__colsample_bytree': [0.6, 0.8, 1.0],
    #'xgbclassifier__gamma': [0, 1, 5],
    #'xgbclassifier__reg_alpha': [0, 1, 10],
    #'xgbclassifier__reg_lambda': [1, 10, 100]
}
    
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=2,scoring='accuracy', verbose=3, n_jobs=-1)
```

```python
grid_search.fit(Xb,yb)
```

```python
grid_search.best_score_
```
